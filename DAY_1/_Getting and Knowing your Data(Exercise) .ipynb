# Ex1 - Getting and Knowing your Data


### Step 1: Initialize PySpark Session



```python
import findspark
findspark.init()
from pyspark.sql import SparkSession

# Create a Spark session
spark = SparkSession.builder.appName("Day1").getOrCreate()
```

### Step 2: Load the Dataset





```python
# Load the Chipotle dataset into a Spark DataFrame
data_path = 'chipotle.csv' # Replace with the actual path
df = spark.read.csv(data_path, header=True, inferSchema=True)
```

### Step 3. Get an overview of the DataFrame:


```python
#this prints out current schema of DataFrame
df.printSchema()
```

```python
#method1: prints out first data
df.show(5)
# #methods2:selects the entire data and shows top 5
# df.select("*").show(5) 
# #method3
# df.selectExpr("*").show(5) 
# #method4:limits the entire data to 5 and shows it
# df.limit(5).show()
```

### Step 4.Calculate basic statistics:


```python
#method1
df.describe().show()
```

```python
#method2: this shows quartiles as well
df.summary().show()
```

### Step 5. What is the number of observations in the dataset?

```python
#count() counts the data in df
print("Number of Observations:", df.count())
```

### Step 6. What is the number of columns in the dataset?

```python
#total length of df columns
print("Number of Columns:", len(df.columns))
```
